Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework for Python. It's designed to be lightweight, flexible, and easy to use, making it a popular choice for building web applications and APIs. Flask provides the basics needed to get a web application up and running, while leaving room for developers to add or customize components according to their specific requirements.

Simplicity: Flask follows a simple and intuitive design, making it easy for developers to understand and work with. It has a small core and minimal built-in functionality, which allows developers to extend it with third-party libraries as needed.

Flexibility: Flask doesn't impose a strict structure on your project. This flexibility means you can choose how to organize your application and which components to use, resulting in a more personalized architecture.

Minimal Boilerplate: Flask reduces the amount of boilerplate code needed to get an application running. This means you can focus more on writing the specific functionality you need rather than dealing with complex setup.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs (Uniform Resource Locators) to specific functions or views in your web application. Each route corresponds to a specific URL path, and when a user accesses that URL, the associated function is executed to generate the appropriate response. Routes are defined using decorators, specifically the @app.route() decorator in Flask.

In Flask, the url_for() function is used for URL building. It helps generate URLs for routes in your application based on the endpoint (name of the function handling the route) and any parameters required by the route.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Welcome to the homepage!'

@app.route('/user/<username>')
def user_profile(username):
    return f'Hello, {username}!'

if __name__ == '__main__':
    app.run(host='0.0.0.0',port=2000)
    with app.test_request_context():
        print(url_for('index'))  # Output: /
        print(url_for('user_profile', username='Alice'))  # Output: /user/Alice
```
