Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
    
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

A bar plot (also known as a bar chart) is a graphical representation used to display categorical data using rectangular bars. Each bar represents a specific category, and the length or height of the bar corresponds to the value associated with that category. Bar plots are commonly used to visualize and compare data across different categories to identify patterns, trends, and differences.

```python
plt.figure()
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("profit")
plt.title("Company profit")
plt.show()
```

```python
plt.figure()
plt.barh(company,profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Company profit")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
    
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)


A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a concise summary of the distribution of a dataset. It displays key statistical measures and highlights potential outliers in the data. Box plots are used to gain insights into the spread, central tendency, and symmetry of data across different groups or categories.

```python
box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90, 20, 200)
plt.figure()

plt.subplot(1,2,1)
plt.boxplot(box1)
plt.title('box plot 1')

plt.subplot(1,2,2)
plt.boxplot(box2)
plt.title('box plot 2')

plt.tight_layout()
plt.show()
```
